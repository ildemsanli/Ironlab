### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(df.country, df.genres, margins=True).All
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab(df.country, [df.director_name, df.genres])
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(df.color, df.duration, values=df.duration, aggfunc='mean', margins=True).All
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
df['Total_likes']=df.actor_1_facebook_likes + df.actor_2_facebook_likes + df.actor_3_facebook_likes

pd.crosstab(df.director_name, df.Total_likes, values=df.Total_likes, aggfunc='mean', margins=True).All
```
